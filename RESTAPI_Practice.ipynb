```
Task 1 — Fetch and Validate the Response
Send a GET request to the endpoint using the requests library inside a try-except block. Print the HTTP status code. If the request is successful (status code 200), print the total number of countries retrieved. If it fails, print a descriptive error message.

```

````python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)
    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries retrieved:", len(data))
    else:
        print("Failed to fetch data. Status code:", response.status_code)
        print("Response:", response.text)

except Exception as e:
    print("An error occurred:", str(e))
````

Task 2 — Extract and Display Key Fields
Parse the JSON response and extract the following fields for each country: common_name, population, region, and area. Store the results in a Pandas DataFrame and display the first 10 rows.

````python
import pandas as pd

countries_list = []

for country in data:
    countries_list.append({
        "common_name": country.get("name", {}).get("common"),
        "population": country.get("population"),
        "region": country.get("region"),
        "area": country.get("area")
    })

df = pd.DataFrame(countries_list)

df.head(10)
````

Task 3 — Summarize the Data
Using the DataFrame from Task 2, answer the following:

Which region has the highest total population?

Which country has the largest area?
Print both answers clearly with labels.

````python
# Region with highest total population
region_population = df.groupby("region")["population"].sum()
highest_population_region = region_population.idxmax()

# Country with largest area
largest_country = df.loc[df["area"].idxmax(), "common_name"]

print("Region with highest total population:", highest_population_region)
print("Country with largest area:", largest_country)
````
